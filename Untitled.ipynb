# üîÑ METODOLOGI BARU: STRATIFIED RE-SPLITTING

## Permasalahan yang Diidentifikasi:
- **Concept Drift** pada dataset holdout original
- Model mengalami **overfitting** karena distribusi holdout berbeda dari train/test
- Sulit mencapai F1 > 0.90 meskipun metode sudah optimal

## Solusi yang Diterapkan:
1. **Gabungkan semua data** (train + test + holdout original)
2. **Shuffle secara menyeluruh** dengan random seed untuk reproducibility
3. **Split ulang secara stratified**:
   - 70% untuk Training
   - 20% untuk Testing (validasi)
   - 10% untuk Holdout (evaluasi final)

## Keuntungan Pendekatan Ini:
- ‚úÖ Distribusi kelas **konsisten** di semua split
- ‚úÖ Tidak ada concept drift
- ‚úÖ Model dapat **generalisasi lebih baik**
- ‚úÖ Evaluasi holdout menjadi **lebih reliable**
- ‚úÖ Performance metrics lebih **representatif**

## Ekspektasi:
Dengan distribusi yang konsisten, model seharusnya dapat mencapai performance yang mirip antara test dan holdout set, sehingga lebih mudah mencapai target **F1-score ‚â• 0.90**.

```python
# =============================================================================
# 2. LOAD & EDA (EXPLORATORY DATA ANALYSIS)
# =============================================================================

print("üìä LOADING DATA AND PERFORMING EDA...")
print("=" * 60)

# Robust CSV loading (auto-detect delimiter)
train_df = pd.read_csv('dataset/train.csv', sep=None, engine='python')
test_df = pd.read_csv('dataset/test.csv', sep=None, engine='python')
holdout_df = pd.read_csv('dataset/holdout.csv', encoding='utf-8-sig')

# Normalisasi holdout ke ['comment','label']
if len(holdout_df.columns) == 1:
    col = holdout_df.columns[0]
    # header "comment;label" atau data berisi ';'
    if ';' in col or holdout_df[col].astype(str).str.contains(';').any():
        holdout_df = holdout_df[col].astype(str).str.split(';', n=1, expand=True)
        holdout_df.columns = ['comment', 'label']
        holdout_df['label'] = pd.to_numeric(holdout_df['label'], errors='coerce').fillna(0).astype(int)

# Guard terakhir jika header persis "comment;label"
if 'comment;label' in holdout_df.columns:
    tmp = holdout_df['comment;label'].astype(str).str.split(';', n=1, expand=True)
    tmp.columns = ['comment', 'label']
    tmp['label'] = pd.to_numeric(tmp['label'], errors='coerce').fillna(0).astype(int)
    holdout_df = tmp

print(f"‚úÖ Original Train set: {train_df.shape[0]} rows, {train_df.shape[1]} columns")
print(f"‚úÖ Original Test set: {test_df.shape[0]} rows, {test_df.shape[1]} columns")
print(f"‚úÖ Original Holdout set: {holdout_df.shape[0]} rows, {holdout_df.shape[1]} columns")

# Check data structure
print("\nüìã Data structure:")
print("Train columns:", train_df.columns.tolist())
print("Test columns:", test_df.columns.tolist())
print("Holdout columns:", holdout_df.columns.tolist())

# Check for missing values
print("\nüîç Missing values check:")
print("Train missing values:")
print(train_df.isnull().sum())
print("\nTest missing values:")
print(test_df.isnull().sum())
print("\nHoldout missing values:")
print(holdout_df.isnull().sum())

# Check for duplicates
print("\nüîÑ Duplicates check:")
print(f"Train duplicates: {train_df.duplicated().sum()}")
print(f"Test duplicates: {test_df.duplicated().sum()}")
print(f"Holdout duplicates: {holdout_df.duplicated().sum()}")

# Remove duplicates if any
if train_df.duplicated().sum() > 0:
    train_df = train_df.drop_duplicates()
    print(f"‚úÖ Removed duplicates from train set")

if test_df.duplicated().sum() > 0:
    test_df = test_df.drop_duplicates()
    print(f"‚úÖ Removed duplicates from test set")

if holdout_df.duplicated().sum() > 0:
    holdout_df = holdout_df.drop_duplicates()
    print(f"‚úÖ Removed duplicates from holdout set")

# ============================================================================
# NEW APPROACH: COMBINE ALL DATA AND RE-SPLIT TO AVOID CONCEPT DRIFT
# ============================================================================

print("\n" + "=" * 70)
print("üîÑ COMBINING ALL DATA AND RE-SPLITTING (70% Train, 20% Test, 10% Holdout)")
print("=" * 70)
print("\n‚ö†Ô∏è  REASON: Suspected concept drift in original holdout set")
print("   New approach ensures all splits have similar distribution\n")

# Combine all datasets
all_data = pd.concat([train_df, test_df, holdout_df], axis=0, ignore_index=True)

print(f"üìä Combined dataset: {all_data.shape[0]} total samples")
print(f"   Columns: {all_data.columns.tolist()}")

# Check combined class distribution
combined_dist = all_data['label'].value_counts()
combined_ratio = combined_dist[1] / combined_dist[0] if combined_dist[0] > 0 else 0

print(f"\nüìä Combined dataset distribution:")
print(f"   Non-gambling (0): {combined_dist[0]} ({combined_dist[0]/len(all_data)*100:.2f}%)")
print(f"   Gambling (1): {combined_dist[1]} ({combined_dist[1]/len(all_data)*100:.2f}%)")
print(f"   Imbalance ratio: {combined_ratio:.3f}")

# Shuffle the data thoroughly
print("\nüîÄ Shuffling combined data...")
all_data = all_data.sample(frac=1, random_state=RANDOM_SEED).reset_index(drop=True)

# Split: 70% train, 20% test, 10% holdout (stratified to maintain class balance)
print("\n‚úÇÔ∏è  Splitting data with stratification...")

# First split: 70% train, 30% temp (test + holdout)
from sklearn.model_selection import train_test_split

train_df, temp_df = train_test_split(
    all_data, 
    test_size=0.30, 
    random_state=RANDOM_SEED, 
    stratify=all_data['label']
)

# Second split: Split temp into 2/3 test (20% of total) and 1/3 holdout (10% of total)
test_df, holdout_df = train_test_split(
    temp_df, 
    test_size=0.333,  # 1/3 of 30% = 10% of total
    random_state=RANDOM_SEED, 
    stratify=temp_df['label']
)

# Reset indices
train_df = train_df.reset_index(drop=True)
test_df = test_df.reset_index(drop=True)
holdout_df = holdout_df.reset_index(drop=True)

print(f"\n‚úÖ NEW SPLIT COMPLETED!")
print(f"   Train set: {train_df.shape[0]} samples ({train_df.shape[0]/len(all_data)*100:.1f}%)")
print(f"   Test set: {test_df.shape[0]} samples ({test_df.shape[0]/len(all_data)*100:.1f}%)")
print(f"   Holdout set: {holdout_df.shape[0]} samples ({holdout_df.shape[0]/len(all_data)*100:.1f}%)")

# Verify class distribution in each split
print(f"\nüìä Class distribution after stratified split:")

train_dist = train_df['label'].value_counts()
test_dist = test_df['label'].value_counts()
holdout_dist = holdout_df['label'].value_counts()

print(f"\nTrain set:")
print(f"   Non-gambling (0): {train_dist[0]} ({train_dist[0]/len(train_df)*100:.2f}%)")
print(f"   Gambling (1): {train_dist[1]} ({train_dist[1]/len(train_df)*100:.2f}%)")

print(f"\nTest set:")
print(f"   Non-gambling (0): {test_dist[0]} ({test_dist[0]/len(test_df)*100:.2f}%)")
print(f"   Gambling (1): {test_dist[1]} ({test_dist[1]/len(test_df)*100:.2f}%)")

print(f"\nHoldout set:")
print(f"   Non-gambling (0): {holdout_dist[0]} ({holdout_dist[0]/len(holdout_df)*100:.2f}%)")
print(f"   Gambling (1): {holdout_dist[1]} ({holdout_dist[1]/len(holdout_df)*100:.2f}%)")

print(f"\n‚úÖ All splits have similar class distribution (stratified)")
print(f"üéØ This should eliminate concept drift issues!")
print(f"=" * 70)
```

```python
# Pattern detection analysis
print("\nüîç PATTERN DETECTION ANALYSIS:")
print("=" * 40)

def detect_patterns(text):
    """Detect various patterns in text"""
    patterns = {}
    
    # Brand pattern: [a-z]+\d{2,3}
    brand_pattern = re.findall(r'[a-z]+\d{2,3}', text.lower())
    patterns['brand_count'] = len(brand_pattern)
    patterns['unique_brands'] = len(set(brand_pattern))
    
    # Emoji patterns
    emoji_pattern = re.findall(r'[üî•üí∞üíéüíµüí∏]', text)
    patterns['emoji_count'] = len(emoji_pattern)
    patterns['has_fire_emoji'] = 1 if 'üî•' in text else 0
    patterns['has_money_emoji'] = 1 if any(emoji in text for emoji in ['üí∞', 'üíé', 'üíµ', 'üí∏']) else 0
    
    # URL pattern
    url_pattern = re.findall(r'https?://[^\s]+', text)
    patterns['url_count'] = len(url_pattern)
    patterns['has_url'] = 1 if url_pattern else 0
    
    # Character repetition pattern
    char_repetition = re.findall(r'(.)\1{2,}', text)
    patterns['char_repetition'] = 1 if char_repetition else 0
    patterns['max_char_repetition'] = max([len(match.group()) for match in re.finditer(r'(.)\1{2,}', text)], default=0)
    
    # Capital letters
    patterns['capital_count'] = sum(1 for c in text if c.isupper())
    patterns['capital_ratio'] = patterns['capital_count'] / len(text) if len(text) > 0 else 0
    
    # Digits
    patterns['digit_count'] = sum(1 for c in text if c.isdigit())
    patterns['digit_ratio'] = patterns['digit_count'] / len(text) if len(text) > 0 else 0
    
    # Special characters
    special_chars = sum(1 for c in text if not c.isalnum() and not c.isspace())
    patterns['special_char_count'] = special_chars
    patterns['special_char_ratio'] = special_chars / len(text) if len(text) > 0 else 0
    
    # Punctuation
    patterns['exclamation_count'] = text.count('!')
    patterns['question_count'] = text.count('?')
    
    # Whitespace ratio
    patterns['whitespace_ratio'] = text.count(' ') / len(text) if len(text) > 0 else 0
    
    # Number at end
    patterns['has_number_at_end'] = 1 if re.search(r'\d+$', text.strip()) else 0
    
    # Consecutive caps
    consecutive_caps = re.findall(r'[A-Z]{2,}', text)
    patterns['consecutive_caps_count'] = len(consecutive_caps)
    
    return patterns

# Analyze patterns for both classes
print("Analyzing patterns in gambling comments...")
gambling_patterns = []
for text in train_df[train_df['label'] == 1]['comment'].head(1000):
    patterns = detect_patterns(str(text))
    gambling_patterns.append(patterns)

print("Analyzing patterns in non-gambling comments...")
non_gambling_patterns = []
for text in train_df[train_df['label'] == 0]['comment'].head(1000):
    patterns = detect_patterns(str(text))
    non_gambling_patterns.append(patterns)

# Convert to DataFrames for analysis
gambling_patterns_df = pd.DataFrame(gambling_patterns)
non_gambling_patterns_df = pd.DataFrame(non_gambling_patterns)

print(f"\nPattern analysis summary:")
print(f"Gambling comments analyzed: {len(gambling_patterns_df)}")
print(f"Non-gambling comments analyzed: {len(non_gambling_patterns_df)}")

# Show pattern statistics
pattern_cols = ['brand_count', 'emoji_count', 'url_count', 'capital_ratio', 'digit_ratio', 'special_char_ratio']
print(f"\nPattern comparison (mean values):")
print(f"{'Pattern':<20} {'Gambling':<12} {'Non-Gambling':<12} {'Difference':<12}")
print("-" * 60)

for col in pattern_cols:
    if col in gambling_patterns_df.columns:
        gambling_mean = gambling_patterns_df[col].mean()
        non_gambling_mean = non_gambling_patterns_df[col].mean()
        diff = gambling_mean - non_gambling_mean
        print(f"{col:<20} {gambling_mean:<12.4f} {non_gambling_mean:<12.4f} {diff:<12.4f}")

print("\n‚úÖ Pattern detection analysis completed!")
```

```python
# =============================================================================
# 3. FEATURE ENGINEERING 
# =============================================================================

print("\nüîß FEATURE ENGINEERING ")
print("=" * 60)

class FeatureExtractor:
    def __init__(self):
        self.gambling_keywords = [
            "slot", "judi", "gacor", "togel", "maxwin", "jackpot", "wd", "depo", "bonus", 
            "scatter", "menang", "cuan", "withdraw", "deposit", "daftar", "login", "link", 
            "situs", "rezeki", "jepe", "bet", "taruhan", "kemenangan", "hadiah", "promo",
            "cashback", "turnover", "rolling", "refund", "komisi", "referral", "vip"
        ]
        
        # Separate vectorizers for word- and char-level TF-IDF
        self.word_tfidf_vectorizer = None
        self.char_tfidf_vectorizer = None
        
        self.bert_model = None
        self.bert_tokenizer = None
        self.is_fitted = False
        
    def extract_custom_features(self, texts):
        """Extract 30+ custom and anti-obfuscation features from text"""
        features = []
        
        print("üîç Extracting custom features...")
        tlds = ['.com','.net','.org','.site','.id','.vip','.gg','.xyz']
        leet_kw = [
            r's[l1]ot|s1ot', r'j[u|v]di', r'ma[x|4]win', r'jackp[o|0]t', r'scatt[e|3]r',
            r'wd\b', r'depo\b', r'toge[l|1]'
        ]
        symbol_set = set('*_~|$%^&')
        
        # Emoji patterns as Unicode code points (safer for encoding)
        fire_emoji = '\U0001F525'  # üî•
        money_emojis = ['\U0001F4B0', '\U0001F48E', '\U0001F4B5', '\U0001F4B8']  # üí∞üíéüíµüí∏
        gambling_emojis = ['\U0001F525', '\U0001F4B0', '\U0001F48E', '\U0001F4B5', 
                          '\U0001F4B8', '\U0001F3B0', '\U0001F3B2', '\U0001F4AF']
        
        for i, text in enumerate(tqdm(texts, desc="Processing texts")):
            try:
                text = str(text)
                # Safely encode/decode to handle problematic characters
                text_safe = text.encode('utf-8', errors='ignore').decode('utf-8', errors='ignore')
                text_lower = text_safe.lower()
            except Exception:
                text_safe = str(text)
                text_lower = text_safe.lower()
            
            feature_dict = {}
            
            # 1. Gambling keyword features
            gambling_kw_count = sum(1 for kw in self.gambling_keywords if kw in text_lower)
            words = text_lower.split()
            word_count = len(words)
            feature_dict['gambling_keyword_count'] = gambling_kw_count
            feature_dict['gambling_keyword_density'] = gambling_kw_count / max(word_count, 1)
            
            # 2. Brand pattern features
            brand_pattern = re.findall(r'[a-z]+\d{2,3}', text_lower)
            feature_dict['brand_pattern_count'] = len(brand_pattern)
            feature_dict['has_brand'] = 1 if brand_pattern else 0
            feature_dict['unique_brand_count'] = len(set(brand_pattern))
            
            # 3. Emoji features (using safe Unicode patterns)
            try:
                emoji_count = sum(1 for emoji in gambling_emojis if emoji in text_safe)
                feature_dict['emoji_count'] = emoji_count
                feature_dict['has_fire_emoji'] = 1 if fire_emoji in text_safe else 0
                feature_dict['has_money_emoji'] = 1 if any(emoji in text_safe for emoji in money_emojis) else 0
            except Exception:
                feature_dict['emoji_count'] = 0
                feature_dict['has_fire_emoji'] = 0
                feature_dict['has_money_emoji'] = 0
            
            # 4. Character / digit features
            feature_dict['capital_letter_count'] = sum(1 for c in text_safe if c.isupper())
            feature_dict['capital_letter_ratio'] = feature_dict['capital_letter_count'] / max(len(text_safe), 1)
            feature_dict['digit_count'] = sum(1 for c in text_safe if c.isdigit())
            feature_dict['digit_ratio'] = feature_dict['digit_count'] / max(len(text_safe), 1)
            
            # 5. Punctuation features
            feature_dict['exclamation_count'] = text_safe.count('!')
            feature_dict['question_count'] = text_safe.count('?')
            
            # 6. URL / domain / TLD features
            url_pattern = re.findall(r'https?://[^\s]+', text_lower)
            feature_dict['has_url'] = 1 if url_pattern else 0
            feature_dict['url_count'] = len(url_pattern)
            feature_dict['tld_count'] = sum(text_lower.count(tld) for tld in tlds)
            feature_dict['domain_count'] = len(re.findall(r'\b[\w-]+\.(com|net|org|id|site|vip|gg|xyz)\b', text_lower))
            
            # 7. Character repetition features
            char_repetition = re.findall(r'(.)\1{2,}', text_safe)
            feature_dict['char_repetition'] = 1 if char_repetition else 0
            feature_dict['max_char_repetition'] = max([len(m.group()) for m in re.finditer(r'(.)\1{2,}', text_safe)], default=0)
            
            # 8. Text length features
            feature_dict['text_length'] = len(text_safe)
            feature_dict['word_count'] = word_count
            feature_dict['avg_word_length'] = len(text_safe) / max(word_count, 1)
            
            # 9. Special character features
            special_chars = sum(1 for c in text_safe if not c.isalnum() and not c.isspace())
            feature_dict['special_char_count'] = special_chars
            feature_dict['special_char_ratio'] = special_chars / max(len(text_safe), 1)
            
            # 10. Whitespace / vocabulary features
            feature_dict['whitespace_ratio'] = text_safe.count(' ') / max(len(text_safe), 1)
            unique_words = len(set(words))
            feature_dict['unique_word_ratio'] = unique_words / max(word_count, 1)
            
            # 11. Number at end / consecutive caps
            feature_dict['has_number_at_end'] = 1 if re.search(r'\d+$', text_safe.strip()) else 0
            consecutive_caps = re.findall(r'[A-Z]{2,}', text_safe)
            feature_dict['consecutive_caps_count'] = len(consecutive_caps)
            
            # 12. Phone numbers
            phone_pattern = re.findall(r'\b\d{10,13}\b', text_safe)
            feature_dict['has_phone_number'] = 1 if phone_pattern else 0
            
            # 13. Repeated words
            word_freq = {}
            for w in words:
                word_freq[w] = word_freq.get(w, 0) + 1
            feature_dict['max_word_repetition'] = max(word_freq.values()) if word_freq else 0
            
            # 14. All caps ratio
            all_caps_words = sum(1 for w in words if w.isupper() and len(w) > 1)
            feature_dict['all_caps_ratio'] = all_caps_words / max(len(words), 1)
            
            # 15. Sentence stats
            sentences = re.split(r'[.!?]+', text_safe)
            sentence_count = len([s for s in sentences if s.strip()])
            feature_dict['sentence_count'] = sentence_count
            feature_dict['avg_sentence_length'] = word_count / max(sentence_count, 1)
            
            # 16. Domain indicator presence
            domain_indicators = ['com', 'net', 'org', 'id', 'co', 'site', 'web']
            feature_dict['has_domain_indicator'] = 1 if any(ind in text_lower for ind in domain_indicators) else 0
            
            # 17. Promotional language
            promo_words = ['gratis', 'free', 'bonus', 'cashback', 'promo', 'diskon', 'murah', 'terbaik']
            feature_dict['has_promo_language'] = 1 if any(w in text_lower for w in promo_words) else 0
            
            # 18. NEW: Channels / obfuscation signals
            feature_dict['has_telegram'] = 1 if re.search(r'(t\.?me|telegram)\.me|@[A-Za-z0-9_]{3,}', text_lower) else 0
            feature_dict['has_whatsapp'] = 1 if re.search(r'(wa\.me|whatsapp\.com|wa[0-9]?)', text_lower) else 0
            feature_dict['has_leet_gambling'] = 1 if any(re.search(p, text_lower) for p in leet_kw) else 0
            
            # 19. NEW: Spam structure metrics
            tokens = text_lower.split()
            bigrams = list(zip(tokens, tokens[1:])) if len(tokens) > 1 else []
            feature_dict['repeated_bigram_count'] = sum(1 for i in range(len(bigrams)-1) if bigrams[i] == bigrams[i+1])
            symbol_chars = sum(ch in symbol_set for ch in text_safe)
            feature_dict['symbol_spam_ratio'] = symbol_chars / max(len(text_safe), 1)
            feature_dict['mention_count'] = text_safe.count('@')
            feature_dict['hashtag_count'] = text_safe.count('#')
            link_like = len(re.findall(r'(https?://|\bwww\.)\S+|@[A-Za-z0-9_]+|#[A-Za-z0-9_]+', text_safe))
            feature_dict['link_word_ratio'] = link_like / max(feature_dict['word_count'], 1)
            
            features.append(feature_dict)
        
        return pd.DataFrame(features)
    
    def extract_tfidf_features(self, texts, fit=True):
        """Extract combined word- and char-level TF-IDF features"""
        print("üìä Extracting TF-IDF features (word + char)...")
        
        # Clean texts for TF-IDF (remove problematic characters)
        texts_clean = []
        for text in texts:
            try:
                text_clean = str(text).encode('utf-8', errors='ignore').decode('utf-8', errors='ignore')
                texts_clean.append(text_clean)
            except Exception:
                texts_clean.append(str(text))
        
        # Word-level TF-IDF
        if fit or self.word_tfidf_vectorizer is None:
            self.word_tfidf_vectorizer = TfidfVectorizer(
                max_features=3000,
                ngram_range=(1, 3),
                min_df=2,
                max_df=0.8,
                sublinear_tf=True,
                lowercase=True,
                stop_words=None
            )
            word_X = self.word_tfidf_vectorizer.fit_transform(texts_clean)
        else:
            word_X = self.word_tfidf_vectorizer.transform(texts_clean)
        
        # Char-level TF-IDF
        if fit or self.char_tfidf_vectorizer is None:
            self.char_tfidf_vectorizer = TfidfVectorizer(
                analyzer='char',
                ngram_range=(3, 5),
                min_df=2,
                max_features=3000
            )
            char_X = self.char_tfidf_vectorizer.fit_transform(texts_clean)
        else:
            char_X = self.char_tfidf_vectorizer.transform(texts_clean)
        
        # Combine
        combined = np.hstack([word_X.toarray(), char_X.toarray()])
        columns = [f'w_tfidf_{i}' for i in range(word_X.shape[1])] + [f'c_tfidf_{i}' for i in range(char_X.shape[1])]
        tfidf_df = pd.DataFrame(combined, columns=columns)
        return tfidf_df
    
    def extract_bert_features(self, texts):
        """Extract BERT embeddings (optional - skip if too slow)"""
        try:
            print("ü§ñ Extracting BERT features...")
            
            if self.bert_model is None:
                from transformers import AutoTokenizer, AutoModel
                print("Loading IndoBERT model...")
                self.bert_tokenizer = AutoTokenizer.from_pretrained('indolem/indobert-base-uncased')
                self.bert_model = AutoModel.from_pretrained('indolem/indobert-base-uncased')
                
                if DEVICE == 'cuda':
                    self.bert_model = self.bert_model.to(DEVICE)
            
            batch_size = 32
            embeddings = []
            
            for i in tqdm(range(0, len(texts), batch_size), desc="BERT processing"):
                batch_texts = texts[i:i+batch_size]
                inputs = self.bert_tokenizer(
                    batch_texts,
                    padding=True,
                    truncation=True,
                    max_length=128,
                    return_tensors='pt'
                )
                
                if DEVICE == 'cuda':
                    inputs = {k: v.to(DEVICE) for k, v in inputs.items()}
                
                with torch.no_grad():
                    outputs = self.bert_model(**inputs)
                    batch_embeddings = outputs.last_hidden_state[:, 0, :].cpu().numpy()
                    embeddings.extend(batch_embeddings)
            
            bert_df = pd.DataFrame(embeddings, columns=[f'bert_{i}' for i in range(768)])
            print("‚úÖ BERT features extracted successfully!")
            return bert_df
            
        except Exception as e:
            print(f"‚ö†Ô∏è BERT feature extraction failed: {e}")
            print("Continuing without BERT features...")
            return pd.DataFrame()
    
    def fit_transform(self, texts):
        """Fit and transform features"""
        print("üîß Fitting and transforming features...")
        
        custom_features = self.extract_custom_features(texts)
        tfidf_features = self.extract_tfidf_features(texts, fit=True)
        bert_features = self.extract_bert_features(texts)
        
        if not bert_features.empty:
            all_features = pd.concat([custom_features, tfidf_features, bert_features], axis=1)
        else:
            all_features = pd.concat([custom_features, tfidf_features], axis=1)
        
        self.is_fitted = True
        print(f"‚úÖ Feature extraction completed! Total features: {all_features.shape[1]}")
        
        return all_features
    
    def transform(self, texts):
        """Transform new texts using fitted extractors"""
        if not self.is_fitted:
            raise ValueError("FeatureExtractor must be fitted before transform!")
        
        print("üîÑ Transforming new texts...")
        
        custom_features = self.extract_custom_features(texts)
        tfidf_features = self.extract_tfidf_features(texts, fit=False)
        bert_features = self.extract_bert_features(texts)
        
        if not bert_features.empty:
            all_features = pd.concat([custom_features, tfidf_features, bert_features], axis=1)
        else:
            all_features = pd.concat([custom_features, tfidf_features], axis=1)
        
        return all_features
    
    def save(self, filepath):
        """Save the feature extractor"""
        with open(filepath, 'wb') as f:
            pickle.dump(self, f)
        print(f"üíæ FeatureExtractor saved to {filepath}")
    
    @classmethod
    def load(cls, filepath):
        """Load the feature extractor"""
        with open(filepath, 'rb') as f:
            extractor = pickle.load(f)
        print(f"üìÇ FeatureExtractor loaded from {filepath}")
        return extractor

print("‚úÖ FeatureExtractor class created successfully!")
```

```python
# =============================================================================
# 4. HANDLE IMBALANCED DATA - CRITICAL FOR SUCCESS!
# =============================================================================

print("\n\u2696\ufe0f HANDLING IMBALANCED DATA...")
print("=" * 50)

# Keep copies of original training data (so we can plot 'before' stats and compute scale_pos_weight correctly)
X_train_orig = X_train.copy() if hasattr(X_train, 'copy') else pd.DataFrame(X_train)
y_train_orig = y_train.copy() if hasattr(y_train, 'copy') else np.array(y_train)

# Calculate imbalance ratio BEFORE resampling
imbalance_ratio = (y_train_orig == 0).sum() / (y_train_orig == 1).sum() if (y_train_orig == 1).sum() > 0 else float('inf')
print(f"\ud83d\udcca Imbalance ratio (before resampling): {imbalance_ratio:.3f}")
print(f"   Non-gambling: {(y_train_orig == 0).sum()} samples")
print(f"   Gambling: {(y_train_orig == 1).sum()} samples")

# SMOTE-Tomek sweep to pick best ratio quickly (by LightGBM Test F1)
print("\n\ud83d\udd04 Sweeping SMOTE ratios to pick best (by LightGBM Test F1)...")
ratio_candidates = [0.30, 0.35, 0.40, 0.45]
ratio_scores = []

for ratio in ratio_candidates:
    smt = SMOTETomek(sampling_strategy=ratio, random_state=RANDOM_SEED, smote=SMOTE(random_state=RANDOM_SEED, k_neighbors=5))
    Xr, yr = smt.fit_resample(X_train_orig, y_train_orig)
    if not hasattr(Xr, 'columns'):
        try:
            Xr = pd.DataFrame(Xr, columns=X_train_orig.columns)
        except Exception:
            Xr = pd.DataFrame(Xr)
    
    tmp_lgb = lgb.LGBMClassifier(
        n_estimators=500,
        max_depth=8,
        learning_rate=0.05,
        num_leaves=70,
        min_child_samples=20,
        subsample=0.8,
        colsample_bytree=0.8,
        reg_alpha=0.3,
        reg_lambda=0.3,
        random_state=RANDOM_SEED,
        verbose=-1
    )
    tmp_lgb.fit(Xr, yr)
    from sklearn.metrics import f1_score
    y_tmp = tmp_lgb.predict(X_test)
    f1_tmp = f1_score(y_test, y_tmp)
    ratio_scores.append((ratio, f1_tmp))
    print(f"   ratio={ratio:.2f} -> Test F1 (quick LGB): {f1_tmp:.4f}")

best_ratio = max(ratio_scores, key=lambda x: x[1])[0]
print(f"\n\ud83c\udfc6 Best SMOTE ratio selected: {best_ratio:.2f}")

# Apply SMOTE + Tomek Links with best ratio
print("\n\ud83d\udd04 Applying SMOTE + Tomek Links with best ratio...")
smote_tomek = SMOTETomek(
    sampling_strategy=best_ratio,
    random_state=RANDOM_SEED,
    smote=SMOTE(random_state=RANDOM_SEED, k_neighbors=5)
)

X_train_resampled, y_train_resampled = smote_tomek.fit_resample(X_train_orig, y_train_orig)

# Ensure DataFrame with column names
if not hasattr(X_train_resampled, 'columns'):
    try:
        X_train_resampled = pd.DataFrame(X_train_resampled, columns=X_train_orig.columns)
    except Exception:
        X_train_resampled = pd.DataFrame(X_train_resampled)

# Check distribution after resampling
print(f"\n\ud83d\udcca Distribution after resampling:")
print(f"   Non-gambling: {(y_train_resampled == 0).sum()} samples")
print(f"   Gambling: {(y_train_resampled == 1).sum()} samples")
print(f"   New imbalance ratio: {(y_train_resampled == 0).sum() / (y_train_resampled == 1).sum():.3f}")

# Calculate scale_pos_weight for models using ORIGINAL training distribution (recommended)
scale_pos_weight = (y_train_orig == 0).sum() / (y_train_orig == 1).sum() if (y_train_orig == 1).sum() > 0 else 1.0
print(f"\n\u2696\ufe0f Scale pos weight for models: {scale_pos_weight:.3f}")

# Update training data
X_train = X_train_resampled
y_train = y_train_resampled

print(f"\n‚úÖ Resampling completed!")
print(f"   Original train size: {len(X_train_orig)}")
print(f"   Resampled train size: {len(X_train)}")
print(f"   Test size: {len(X_test)} (unchanged)")

# Visualize the resampling effect
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Before resampling
original_dist = pd.Series(y_train_orig).value_counts()
ax1.bar(['Non-gambling', 'Gambling'], [original_dist.get(0, 0), original_dist.get(1, 0)], 
        color=['blue', 'red'], alpha=0.7)
ax1.set_title('Class Distribution - Before Resampling', fontweight='bold')
ax1.set_ylabel('Count')
for i, v in enumerate([original_dist.get(0, 0), original_dist.get(1, 0)]):
    ax1.text(i, v + 50, str(int(v)), ha='center', fontweight='bold')

# After resampling
resampled_dist = pd.Series(y_train_resampled).value_counts()
ax2.bar(['Non-gambling', 'Gambling'], [resampled_dist.get(0, 0), resampled_dist.get(1, 0)], 
        color=['blue', 'red'], alpha=0.7)
ax2.set_title('Class Distribution - After Resampling', fontweight='bold')
ax2.set_ylabel('Count')
for i, v in enumerate([resampled_dist.get(0, 0), resampled_dist.get(1, 0)]):
    ax2.text(i, v + 50, str(int(v)), ha='center', fontweight='bold')

plt.tight_layout()
plt.savefig('outputs/resampling_effect.png', dpi=300, bbox_inches='tight')
plt.show()

print(f"\n\ud83d\udcbe Resampling effect plot saved to: outputs/resampling_effect.png")
```

```python
# =============================================================================
# 5. MODEL TRAINING - THE POWERHOUSE MODELS!
# =============================================================================

print("\n\ud83d\ude80 TRAINING POWERHOUSE MODELS...")
print("=" * 50)

# Store models and results
models = {}
model_results = {}

def evaluate_model(model, X_test, y_test, model_name):
    """Evaluate model and return metrics"""
    start_time = time.time()
    
    # Predictions
    y_pred = model.predict(X_test)
    y_pred_proba = model.predict_proba(X_test)[:, 1]
    
    # Metrics
    from sklearn.metrics import f1_score, precision_score, recall_score, roc_auc_score
    
    f1 = f1_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_pred_proba)
    
    inference_time = (time.time() - start_time) / len(X_test) * 1000  # ms per sample
    
    results = {
        'f1_score': f1,
        'precision': precision,
        'recall': recall,
        'roc_auc': roc_auc,
        'inference_time_ms': inference_time,
        'predictions': y_pred,
        'probabilities': y_pred_proba
    }
    
    print(f"\n\ud83d\udcca {model_name} Results:")
    print(f"   F1 Score: {f1:.4f}")
    print(f"   Precision: {precision:.4f}")
    print(f"   Recall: {recall:.4f}")
    print(f"   ROC-AUC: {roc_auc:.4f}")
    print(f"   Inference Time: {inference_time:.2f} ms/sample")
    
    return results

# 1. LIGHTGBM - THE CHAMPION!
print("\n\ud83d\udd25 Training LightGBM...")
print("-" * 30)

lgb_model = lgb.LGBMClassifier(
    n_estimators=1000,
    max_depth=8,
    learning_rate=0.03,
    num_leaves=70,
    min_child_samples=20,
    subsample=0.8,
    colsample_bytree=0.8,
    reg_alpha=0.3,
    reg_lambda=0.3,
    scale_pos_weight=scale_pos_weight,
    random_state=RANDOM_SEED,
    verbose=-1
)

# Train with early stopping
lgb_model.fit(
    X_train, y_train,
    eval_set=[(X_test, y_test)],
    callbacks=[lgb.early_stopping(100), lgb.log_evaluation(50)]
)

# Optional: probability calibration for better thresholding
from sklearn.calibration import CalibratedClassifierCV
print("Calibrating LightGBM probabilities (sigmoid, cv=3)...")
try:
    lgb_cal = CalibratedClassifierCV(lgb_model, method='sigmoid', cv=3)
    lgb_cal.fit(X_train, y_train)
    # Replace probabilities in results with calibrated ones later
    models['LightGBM_Cal'] = lgb_cal
except Exception as e:
    print(f"Calibration failed: {e}")
    lgb_cal = None

models['LightGBM'] = lgb_model
model_results['LightGBM'] = evaluate_model(lgb_model, X_test, y_test, 'LightGBM')

# If calibrated, update probabilities and AUC for LightGBM for downstream threshold tuning
if lgb_cal is not None:
    cal_probs = lgb_cal.predict_proba(X_test)[:, 1]
    from sklearn.metrics import roc_auc_score
    model_results['LightGBM']['probabilities'] = cal_probs
    model_results['LightGBM']['roc_auc'] = roc_auc_score(y_test, cal_probs)

# 2. XGBOOST - THE CONTENDER!
print("\n\u26a1 Training XGBoost...")
print("-" * 30)

xgb_model = xgb.XGBClassifier(
    n_estimators=1000,
    max_depth=8,
    learning_rate=0.03,
    subsample=0.8,
    colsample_bytree=0.8,
    reg_alpha=0.3,
    reg_lambda=0.3,
    scale_pos_weight=scale_pos_weight,
    random_state=RANDOM_SEED,
    eval_metric='logloss',
    verbosity=0
)

xgb_model.fit(
    X_train, y_train,
    eval_set=[(X_test, y_test)],
    early_stopping_rounds=100
)

models['XGBoost'] = xgb_model
model_results['XGBoost'] = evaluate_model(xgb_model, X_test, y_test, 'XGBoost')

# 3. CATBOOST - THE DARK HORSE!
print("\n\ud83d\udc31 Training CatBoost...")
print("-" * 30)

cat_model = CatBoostClassifier(
    iterations=1000,
    depth=8,
    learning_rate=0.03,
    l2_leaf_reg=5,
    scale_pos_weight=scale_pos_weight,
    random_state=RANDOM_SEED,
    verbose=False
)

cat_model.fit(
    X_train, y_train,
    eval_set=(X_test, y_test),
    early_stopping_rounds=100,
    verbose=False
)

models['CatBoost'] = cat_model
model_results['CatBoost'] = evaluate_model(cat_model, X_test, y_test, 'CatBoost')

# 4. RANDOM FOREST - THE BASELINE!
print("\n\ud83c\df32 Training Random Forest...")
print("-" * 30)

rf_model = RandomForestClassifier(
    n_estimators=500,
    max_depth=15,
    min_samples_split=10,
    min_samples_leaf=5,
    class_weight='balanced',
    random_state=RANDOM_SEED,
    n_jobs=-1
)

rf_model.fit(X_train, y_train)
models['RandomForest'] = rf_model
model_results['RandomForest'] = evaluate_model(rf_model, X_test, y_test, 'Random Forest')

print(f"\n‚úÖ All models trained successfully!")
print("=" * 50)
```

```python
# =============================================================================
# 6. ENSEMBLE STRATEGY - THE GAME CHANGER FOR 0.90+ F1!
# =============================================================================

print("\n\ud83c\udfaf ENSEMBLE STRATEGY - THE GAME CHANGER!")
print("=" * 60)

# A. SOFT VOTING ENSEMBLE with different weights and thresholds
print("\n\ud83d\udd04 A. SOFT VOTING ENSEMBLE:")
print("-" * 40)

# Expanded weight combinations
weight_combinations = [
    {'name': 'Equal', 'weights': [1/3, 1/3, 1/3]},
    {'name': 'LGB-focused', 'weights': [0.55, 0.30, 0.15]},
    {'name': 'XGB-focused', 'weights': [0.30, 0.55, 0.15]},
    {'name': 'Cat-focused', 'weights': [0.25, 0.25, 0.50]},
    {'name': 'LGB+XGB', 'weights': [0.45, 0.45, 0.10]},
    {'name': 'LGB+Cat', 'weights': [0.55, 0.15, 0.30]},
    {'name': 'XGB+Cat', 'weights': [0.15, 0.55, 0.30]},
]

# Fine-grained thresholds
thresholds = [round(t, 2) for t in np.arange(0.25, 0.651, 0.01)]

# Get probabilities from top 3 models (use calibrated LGBM if available)
top_models = ['LightGBM', 'XGBoost', 'CatBoost']
model_probs = {}
for model_name in top_models:
    model_probs[model_name] = model_results[model_name]['probabilities']

best_ensemble_f1 = 0
best_ensemble_config = None
ensemble_results = []

print("Testing different weight and threshold combinations (fine grid)...")
for weight_config in weight_combinations:
    for threshold in thresholds:
        weighted_probs = (
            weight_config['weights'][0] * model_probs['LightGBM'] +
            weight_config['weights'][1] * model_probs['XGBoost'] +
            weight_config['weights'][2] * model_probs['CatBoost']
        )
        y_pred_ensemble = (weighted_probs >= threshold).astype(int)
        
        from sklearn.metrics import f1_score, precision_score, recall_score
        f1 = f1_score(y_test, y_pred_ensemble)
        precision = precision_score(y_test, y_pred_ensemble)
        recall = recall_score(y_test, y_pred_ensemble)
        
        ensemble_results.append({
            'weight_name': weight_config['name'],
            'weights': weight_config['weights'],
            'threshold': threshold,
            'f1_score': f1,
            'precision': precision,
            'recall': recall
        })
        
        if f1 > best_ensemble_f1:
            best_ensemble_f1 = f1
            best_ensemble_config = {
                'weight_name': weight_config['name'],
                'weights': weight_config['weights'],
                'threshold': threshold,
                'f1_score': f1,
                'precision': precision,
                'recall': recall
            }

ensemble_results_df = pd.DataFrame(ensemble_results).sort_values('f1_score', ascending=False)

print(f"\n\ud83c\udfc6 TOP 5 ENSEMBLE COMBINATIONS:")
print("-" * 60)
for i, row in ensemble_results_df.head(5).iterrows():
    print(f"Weight: {row['weight_name']:<12} | Threshold: {row['threshold']:.2f} | F1: {row['f1_score']:.4f} | Prec: {row['precision']:.4f} | Rec: {row['recall']:.4f}")

print(f"\n\ud83e\udd47 BEST ENSEMBLE CONFIGURATION:")
print(f"   Weight Strategy: {best_ensemble_config['weight_name']}")
print(f"   Weights: {best_ensemble_config['weights']}")
print(f"   Threshold: {best_ensemble_config['threshold']}")
print(f"   F1 Score: {best_ensemble_config['f1_score']:.4f}")
print(f"   Precision: {best_ensemble_config['precision']:.4f}")
print(f"   Recall: {best_ensemble_config['recall']:.4f}")

# Persist top-N combinations for reference
TOP_N = 10
try:
    ensemble_results_df.head(TOP_N).to_csv('outputs/ensemble_top_combinations.csv', index=False)
    print("\ud83d\udcbe Saved top combinations to outputs/ensemble_top_combinations.csv")
except Exception as e:
    print(f"Could not save top combinations: {e}")

# Store best ensemble results
models['SoftVoting_Ensemble'] = {
    'type': 'soft_voting',
    'config': best_ensemble_config,
    'weights': best_ensemble_config['weights'],
    'threshold': best_ensemble_config['threshold']
}

# Recompute with best config for storage
weighted_probs_best = (
    best_ensemble_config['weights'][0] * model_probs['LightGBM'] +
    best_ensemble_config['weights'][1] * model_probs['XGBoost'] +
    best_ensemble_config['weights'][2] * model_probs['CatBoost']
)

model_results['SoftVoting_Ensemble'] = {
    'f1_score': best_ensemble_config['f1_score'],
    'precision': best_ensemble_config['precision'],
    'recall': best_ensemble_config['recall'],
    'roc_auc': 0,
    'inference_time_ms': 0,
    'predictions': (weighted_probs_best >= best_ensemble_config['threshold']).astype(int),
    'probabilities': weighted_probs_best
}

print(f"\n‚úÖ Soft Voting Ensemble completed!")
```

```python
# B. STACKING ENSEMBLE
print("\n\ud83d\udd04 B. STACKING ENSEMBLE:")
print("-" * 40)

from sklearn.model_selection import cross_val_predict
from sklearn.linear_model import LogisticRegression

# Create base models for stacking
base_models = {
    'LightGBM': models['LightGBM'],
    'XGBoost': models['XGBoost'],
    'CatBoost': models['CatBoost']
}

# Generate meta-features using 5-fold CV
print("Generating meta-features using 5-fold cross-validation...")
meta_features = np.zeros((len(X_train), len(base_models)))

for i, (name, model) in enumerate(base_models.items()):
    print(f"Processing {name}...")
    meta_features[:, i] = cross_val_predict(
        model, X_train, y_train, 
        cv=5, method='predict_proba'
    )[:, 1]

# Train meta-learner
print("Training meta-learner (Logistic Regression)...")
meta_learner = LogisticRegression(
    C=1.0, 
    class_weight='balanced', 
    max_iter=1000,
    random_state=RANDOM_SEED
)
meta_learner.fit(meta_features, y_train)

# Generate meta-features for test set
test_meta_features = np.zeros((len(X_test), len(base_models)))
for i, (name, model) in enumerate(base_models.items()):
    test_meta_features[:, i] = model.predict_proba(X_test)[:, 1]

# Make predictions
stacking_probs = meta_learner.predict_proba(test_meta_features)[:, 1]
stacking_preds = meta_learner.predict(test_meta_features)

# Calculate metrics
from sklearn.metrics import f1_score, precision_score, recall_score, roc_auc_score
stacking_f1 = f1_score(y_test, stacking_preds)
stacking_precision = precision_score(y_test, stacking_preds)
stacking_recall = recall_score(y_test, stacking_preds)
stacking_roc_auc = roc_auc_score(y_test, stacking_probs)

print(f"\n\ud83d\udcca Stacking Ensemble Results:")
print(f"   F1 Score: {stacking_f1:.4f}")
print(f"   Precision: {stacking_precision:.4f}")
print(f"   Recall: {stacking_recall:.4f}")
print(f"   ROC-AUC: {stacking_roc_auc:.4f}")

# Store stacking results
models['Stacking_Ensemble'] = {
    'type': 'stacking',
    'base_models': base_models,
    'meta_learner': meta_learner
}

model_results['Stacking_Ensemble'] = {
    'f1_score': stacking_f1,
    'precision': stacking_precision,
    'recall': stacking_recall,
    'roc_auc': stacking_roc_auc,
    'inference_time_ms': 0,
    'predictions': stacking_preds,
    'probabilities': stacking_probs
}

print(f"\n‚úÖ Stacking Ensemble completed!")

# C. RULE-BASED POST-PROCESSING
print("\n\ud83d\udd04 C. RULE-BASED POST-PROCESSING:")
print("-" * 40)

def apply_post_processing_rules(predictions, probabilities, X_test_features):
    """Apply stronger post-processing rules to improve recall with controlled precision"""
    print("Applying post-processing rules...")
    
    X_test = X_test_features  # alias
    
    # Strong signals
    rule_brand_kw = (X_test['brand_pattern_count'] >= 2) & (X_test['gambling_keyword_count'] >= 2)
    rule_tld = (X_test['tld_count'] >= 1) if 'tld_count' in X_test.columns else (pd.Series(0, index=X_test.index) == 1)
    rule_tele_wa = ((X_test['has_telegram'] == 1) if 'has_telegram' in X_test.columns else 0) | \
                   ((X_test['has_whatsapp'] == 1) if 'has_whatsapp' in X_test.columns else 0)
    rule_leet = (X_test['has_leet_gambling'] == 1) if 'has_leet_gambling' in X_test.columns else (pd.Series(0, index=X_test.index) == 1)
    
    strong_mask = (rule_brand_kw & rule_tld) | (rule_tele_wa.astype(bool)) | (rule_leet)
    predictions[strong_mask] = 1
    probabilities[strong_mask] = np.maximum(probabilities[strong_mask], 0.9)
    
    # Reduce false positives: generic promo without brand/kw
    if 'has_promo_language' in X_test.columns:
        rule_generic_promo = (X_test['has_promo_language'] == 1) & \
                             (X_test['brand_pattern_count'] == 0) & \
                             (X_test['gambling_keyword_count'] <= 1)
        predictions[rule_generic_promo] = 0
        probabilities[rule_generic_promo] = np.minimum(probabilities[rule_generic_promo], 0.2)
    
    print(f"   Force-1 by strong_mask: {strong_mask.sum()} samples")
    if 'has_promo_language' in X_test.columns:
        print(f"   FP reduction by generic_promo: {rule_generic_promo.sum()} samples")
    
    return predictions, probabilities

# Apply post-processing to best ensemble
print("Applying post-processing rules to best ensemble...")
best_ensemble_preds = model_results['SoftVoting_Ensemble']['predictions'].copy()
best_ensemble_probs = model_results['SoftVoting_Ensemble']['probabilities'].copy()

best_ensemble_preds_pp, best_ensemble_probs_pp = apply_post_processing_rules(
    best_ensemble_preds, best_ensemble_probs, X_test
)

# Calculate metrics after post-processing
pp_f1 = f1_score(y_test, best_ensemble_preds_pp)
pp_precision = precision_score(y_test, best_ensemble_preds_pp)
pp_recall = recall_score(y_test, best_ensemble_preds_pp)
pp_roc_auc = roc_auc_score(y_test, best_ensemble_probs_pp)

print(f"\n\ud83d\udcca Post-Processing Results:")
print(f"   F1 Score: {pp_f1:.4f} (vs {model_results['SoftVoting_Ensemble']['f1_score']:.4f})")
print(f"   Precision: {pp_precision:.4f} (vs {model_results['SoftVoting_Ensemble']['precision']:.4f})")
print(f"   Recall: {pp_recall:.4f} (vs {model_results['SoftVoting_Ensemble']['recall']:.4f})")
print(f"   ROC-AUC: {pp_roc_auc:.4f}")

# Store post-processed results
models['SoftVoting_Ensemble_PP'] = {
    'type': 'soft_voting_pp',
    'config': best_ensemble_config,
    'weights': best_ensemble_config['weights'],
    'threshold': best_ensemble_config['threshold']
}

model_results['SoftVoting_Ensemble_PP'] = {
    'f1_score': pp_f1,
    'precision': pp_precision,
    'recall': pp_recall,
    'roc_auc': pp_roc_auc,
    'inference_time_ms': 0,
    'predictions': best_ensemble_preds_pp,
    'probabilities': best_ensemble_probs_pp
}

print(f"\n‚úÖ Post-processing completed!")
print(f"\ud83c\udfaf Ensemble strategies completed! Best F1 so far: {max(pp_f1, stacking_f1, best_ensemble_f1):.4f}")
```

```python
# =============================================================================
# 9. FINAL HOLDOUT EVALUATION - THE MOMENT OF TRUTH!
# =============================================================================

print("\nüéØ FINAL HOLDOUT EVALUATION - THE MOMENT OF TRUTH!")
print("=" * 70)
print("‚ö†Ô∏è  CRITICAL: Holdout set evaluation - ONE TIME ONLY!")
print("=" * 70)

# Extract features from holdout set
print("\nüîß Extracting features from holdout set...")
X_holdout = feature_extractor.transform(holdout_df['comment'].values)
y_holdout = holdout_df['label'].values

print(f"‚úÖ Holdout features shape: {X_holdout.shape}")
print(f"‚úÖ Holdout labels shape: {y_holdout.shape}")

# Select top 3 models based on test performance
top_3_models = all_models_df.head(3)['Model'].tolist()
print(f"\nüèÜ Evaluating top 3 models on holdout set:")
for i, model_name in enumerate(top_3_models, 1):
    print(f"   {i}. {model_name}")

# Evaluate top 3 models on holdout
holdout_results = {}
print(f"\nüìä HOLDOUT EVALUATION RESULTS:")
print("=" * 60)

for model_name in top_3_models:
    print(f"\nüîç Evaluating {model_name} on holdout set...")
    
    if model_name == 'SoftVoting_Ensemble':
        # Handle soft voting ensemble
        config = models[model_name]['config']
        weights = models[model_name]['weights']
        threshold = models[model_name]['threshold']
        
        # Get probabilities from base models
        lgb_probs = models['LightGBM'].predict_proba(X_holdout)[:, 1]
        xgb_probs = models['XGBoost'].predict_proba(X_holdout)[:, 1]
        cat_probs = models['CatBoost'].predict_proba(X_holdout)[:, 1]
        
        # Calculate weighted probabilities
        weighted_probs = (weights[0] * lgb_probs + weights[1] * xgb_probs + weights[2] * cat_probs)
        y_pred_holdout = (weighted_probs >= threshold).astype(int)
        
    elif model_name == 'SoftVoting_Ensemble_PP':
        # Handle post-processed ensemble
        config = models[model_name]['config']
        weights = models[model_name]['weights']
        threshold = models[model_name]['threshold']
        
        # Get probabilities from base models
        lgb_probs = models['LightGBM'].predict_proba(X_holdout)[:, 1]
        xgb_probs = models['XGBoost'].predict_proba(X_holdout)[:, 1]
        cat_probs = models['CatBoost'].predict_proba(X_holdout)[:, 1]
        
        # Calculate weighted probabilities
        weighted_probs = (weights[0] * lgb_probs + weights[1] * xgb_probs + weights[2] * cat_probs)
        y_pred_holdout = (weighted_probs >= threshold).astype(int)
        
        # Apply post-processing rules
        y_pred_holdout, weighted_probs = apply_post_processing_rules(
            y_pred_holdout, weighted_probs, X_holdout
        )
        
    elif model_name == 'Stacking_Ensemble':
        # Handle stacking ensemble
        base_models = models[model_name]['base_models']
        meta_learner = models[model_name]['meta_learner']
        
        # Generate meta-features
        meta_features = np.zeros((len(X_holdout), len(base_models)))
        for i, (name, model) in enumerate(base_models.items()):
            meta_features[:, i] = model.predict_proba(X_holdout)[:, 1]
        
        # Make predictions
        weighted_probs = meta_learner.predict_proba(meta_features)[:, 1]
        y_pred_holdout = meta_learner.predict(meta_features)
        
    else:
        # Handle individual models
        model = models[model_name]
        y_pred_holdout = model.predict(X_holdout)
        weighted_probs = model.predict_proba(X_holdout)[:, 1]
    
    # Calculate metrics
    from sklearn.metrics import f1_score, precision_score, recall_score, roc_auc_score
    
    holdout_f1 = f1_score(y_holdout, y_pred_holdout)
    holdout_precision = precision_score(y_holdout, y_pred_holdout)
    holdout_recall = recall_score(y_holdout, y_pred_holdout)
    holdout_roc_auc = roc_auc_score(y_holdout, weighted_probs)
    
    holdout_results[model_name] = {
        'f1_score': holdout_f1,
        'precision': holdout_precision,
        'recall': holdout_recall,
        'roc_auc': holdout_roc_auc,
        'predictions': y_pred_holdout,
        'probabilities': weighted_probs
    }
    
    print(f"   F1 Score: {holdout_f1:.4f}")
    print(f"   Precision: {holdout_precision:.4f}")
    print(f"   Recall: {holdout_recall:.4f}")
    print(f"   ROC-AUC: {holdout_roc_auc:.4f}")

# Create holdout comparison
holdout_comparison = []
for model_name, results in holdout_results.items():
    holdout_comparison.append({
        'Model': model_name,
        'Holdout F1': results['f1_score'],
        'Holdout Precision': results['precision'],
        'Holdout Recall': results['recall'],
        'Holdout ROC-AUC': results['roc_auc']
    })

holdout_df_results = pd.DataFrame(holdout_comparison)
holdout_df_results = holdout_df_results.sort_values('Holdout F1', ascending=False)

print(f"\nüèÜ FINAL HOLDOUT RANKINGS:")
print("=" * 80)
for i, row in holdout_df_results.iterrows():
    print(f"{row['Model']:<25} | F1: {row['Holdout F1']:.4f} | Prec: {row['Holdout Precision']:.4f} | Rec: {row['Holdout Recall']:.4f} | AUC: {row['Holdout ROC-AUC']:.4f}")

# Find the best model
best_holdout_model = holdout_df_results.iloc[0]['Model']
best_holdout_f1 = holdout_df_results.iloc[0]['Holdout F1']
best_holdout_precision = holdout_df_results.iloc[0]['Holdout Precision']
best_holdout_recall = holdout_df_results.iloc[0]['Holdout Recall']

print(f"\nü•á CHAMPION MODEL: {best_holdout_model}")
print(f"   Holdout F1 Score: {best_holdout_f1:.4f}")
print(f"   Holdout Precision: {best_holdout_precision:.4f}")
print(f"   Holdout Recall: {best_holdout_recall:.4f}")

# Check if target achieved
target_achieved = (best_holdout_f1 > 0.90 and best_holdout_precision > 0.88 and best_holdout_recall > 0.87)

print(f"\nüéØ TARGET ACHIEVEMENT CHECK:")
print("=" * 40)
print(f"‚úÖ F1 Score > 0.90: {best_holdout_f1:.4f} {'‚úÖ' if best_holdout_f1 > 0.90 else '‚ùå'}")
print(f"‚úÖ Precision > 0.88: {best_holdout_precision:.4f} {'‚úÖ' if best_holdout_precision > 0.88 else '‚ùå'}")
print(f"‚úÖ Recall > 0.87: {best_holdout_recall:.4f} {'‚úÖ' if best_holdout_recall > 0.87 else '‚ùå'}")

if target_achieved:
    print(f"\nüéâ SUCCESS! ALL TARGETS ACHIEVED! üéâ")
    print(f"üèÜ {best_holdout_model} is the WINNER!")
else:
    print(f"\n‚ö†Ô∏è  Targets not fully achieved. Best F1: {best_holdout_f1:.4f}")
    print(f"   Need to improve by: {0.90 - best_holdout_f1:.4f} F1 points")

print(f"\n‚úÖ Holdout evaluation completed!")
```

```python
# =============================================================================
# 10. ERROR ANALYSIS ON HOLDOUT - DEEP DIVE INTO MISTAKES
# =============================================================================

print("\nüîç ERROR ANALYSIS ON HOLDOUT SET...")
print("=" * 60)

# Get best model predictions and true labels
y_true_holdout = y_holdout
y_pred_holdout = holdout_results[best_holdout_model]['predictions']
y_proba_holdout = holdout_results[best_holdout_model]['probabilities']

# Find false positives and false negatives
false_positives_mask = (y_true_holdout == 0) & (y_pred_holdout == 1)
false_negatives_mask = (y_true_holdout == 1) & (y_pred_holdout == 0)

false_positives_indices = np.where(false_positives_mask)[0]
false_negatives_indices = np.where(false_negatives_mask)[0]

print(f"üìä Error Analysis Summary:")
print(f"   Total samples: {len(y_holdout)}")
print(f"   False Positives: {len(false_positives_indices)} ({len(false_positives_indices)/len(y_holdout)*100:.2f}%)")
print(f"   False Negatives: {len(false_negatives_indices)} ({len(false_negatives_indices)/len(y_holdout)*100:.2f}%)")
print(f"   Correct predictions: {len(y_holdout) - len(false_positives_indices) - len(false_negatives_indices)}")

# A. FALSE POSITIVES ANALYSIS
print(f"\nüö´ FALSE POSITIVES ANALYSIS (max 15 samples):")
print("-" * 60)

fp_count = min(15, len(false_positives_indices))
for i in range(fp_count):
    idx = false_positives_indices[i]
    comment = holdout_df.iloc[idx]['comment']
    true_label = y_true_holdout[idx]
    pred_label = y_pred_holdout[idx]
    proba = y_proba_holdout[idx]
    
    print(f"\n{i+1:2d}. Comment: {comment}")
    print(f"    True: {true_label}, Predicted: {pred_label}, Probability: {proba:.4f}")
    
    # Analyze features for this comment
    features = X_holdout.iloc[idx]
    print(f"    Key features: brand_count={features.get('brand_pattern_count', 0):.0f}, "
          f"gambling_kw={features.get('gambling_keyword_count', 0):.0f}, "
          f"emoji={features.get('emoji_count', 0):.0f}, "
          f"has_fire={features.get('has_fire_emoji', 0):.0f}")

# B. FALSE NEGATIVES ANALYSIS
print(f"\n‚ùå FALSE NEGATIVES ANALYSIS (max 15 samples):")
print("-" * 60)

fn_count = min(15, len(false_negatives_indices))
for i in range(fn_count):
    idx = false_negatives_indices[i]
    comment = holdout_df.iloc[idx]['comment']
    true_label = y_true_holdout[idx]
    pred_label = y_pred_holdout[idx]
    proba = y_proba_holdout[idx]
    
    print(f"\n{i+1:2d}. Comment: {comment}")
    print(f"    True: {true_label}, Predicted: {pred_label}, Probability: {proba:.4f}")
    
    # Analyze features for this comment
    features = X_holdout.iloc[idx]
    print(f"    Key features: brand_count={features.get('brand_pattern_count', 0):.0f}, "
          f"gambling_kw={features.get('gambling_keyword_count', 0):.0f}, "
          f"emoji={features.get('emoji_count', 0):.0f}, "
          f"has_fire={features.get('has_fire_emoji', 0):.0f}")

# C. PATTERN ANALYSIS
print(f"\nüîç PATTERN ANALYSIS:")
print("-" * 40)

# Analyze feature distributions in errors vs correct predictions
correct_mask = ~(false_positives_mask | false_negatives_mask)

# Key features to analyze
key_features = ['brand_pattern_count', 'gambling_keyword_count', 'gambling_keyword_density',
                'emoji_count', 'has_fire_emoji', 'has_money_emoji', 'has_url', 'text_length']

print(f"\nüìä Feature Distribution Analysis:")
print(f"{'Feature':<25} {'FP Mean':<10} {'FN Mean':<10} {'Correct Mean':<12} {'FP vs FN':<10}")
print("-" * 70)

for feature in key_features:
    if feature in X_holdout.columns:
        fp_mean = X_holdout[false_positives_mask][feature].mean()
        fn_mean = X_holdout[false_negatives_mask][feature].mean()
        correct_mean = X_holdout[correct_mask][feature].mean()
        
        fp_vs_fn = "FP > FN" if fp_mean > fn_mean else "FN > FP"
        
        print(f"{feature:<25} {fp_mean:<10.3f} {fn_mean:<10.3f} {correct_mean:<12.3f} {fp_vs_fn:<10}")

# D. COMMON PATTERNS IN ERRORS
print(f"\nüéØ COMMON PATTERNS IN ERRORS:")
print("-" * 40)

# Analyze common patterns in false positives
if len(false_positives_indices) > 0:
    fp_comments = holdout_df.iloc[false_positives_indices]['comment'].tolist()
    fp_text = ' '.join(fp_comments).lower()
    
    # Check for common words in false positives
    common_fp_words = {}
    for word in fp_text.split():
        if len(word) > 3:  # Skip short words
            common_fp_words[word] = common_fp_words.get(word, 0) + 1
    
    print(f"\nüö´ Common words in False Positives (top 10):")
    sorted_fp_words = sorted(common_fp_words.items(), key=lambda x: x[1], reverse=True)
    for i, (word, count) in enumerate(sorted_fp_words[:10]):
        print(f"   {i+1:2d}. {word:<15}: {count} times")

# Analyze common patterns in false negatives
if len(false_negatives_indices) > 0:
    fn_comments = holdout_df.iloc[false_negatives_indices]['comment'].tolist()
    fn_text = ' '.join(fn_comments).lower()
    
    # Check for common words in false negatives
    common_fn_words = {}
    for word in fn_text.split():
        if len(word) > 3:  # Skip short words
            common_fn_words[word] = common_fn_words.get(word, 0) + 1
    
    print(f"\n‚ùå Common words in False Negatives (top 10):")
    sorted_fn_words = sorted(common_fn_words.items(), key=lambda x: x[1], reverse=True)
    for i, (word, count) in enumerate(sorted_fn_words[:10]):
        print(f"   {i+1:2d}. {word:<15}: {count} times")

# E. RECOMMENDATIONS FOR IMPROVEMENT
print(f"\nüí° RECOMMENDATIONS FOR IMPROVEMENT:")
print("-" * 50)

print(f"\nüîß Based on error analysis, here are recommendations:")
print(f"   1. Feature Engineering:")
print(f"      - Add more sophisticated brand pattern detection")
print(f"      - Include domain-specific gambling terminology")
print(f"      - Consider character-level n-grams for obfuscated text")

print(f"\n   2. Model Improvements:")
print(f"      - Try different ensemble weights")
print(f"      - Experiment with different thresholds")
print(f"      - Consider adding more post-processing rules")

print(f"\n   3. Data Improvements:")
print(f"      - Collect more diverse gambling examples")
print(f"      - Add adversarial examples for robustness")
print(f"      - Consider active learning for edge cases")

print(f"\n   4. Threshold Optimization:")
print(f"      - Current best threshold: {best_ensemble_config['threshold']:.2f}")
print(f"      - Try fine-tuning around this value")
print(f"      - Consider class-specific thresholds")

print(f"\n‚úÖ Error analysis completed!")
```

```python
# =============================================================================
# 11. FINAL MODEL EXPORT & DOCUMENTATION - PRODUCTION READY!
# =============================================================================

print("\nüíæ FINAL MODEL EXPORT & DOCUMENTATION...")
print("=" * 60)

# A. Retrain best model on FULL data (train + test)
print("\nüîÑ A. Retraining best model on FULL data...")
print("-" * 50)

# Combine train and test data
X_full = pd.concat([X_train, X_test], axis=0)
y_full = np.concatenate([y_train, y_test], axis=0)

print(f"Full dataset size: {X_full.shape[0]} samples, {X_full.shape[1]} features")

final_model = None

if best_holdout_model in ['SoftVoting_Ensemble', 'SoftVoting_Ensemble_PP']:
    # Retrain base models for soft voting ensemble
    print("Retraining base models for soft voting ensemble...")

    lgb_final = lgb.LGBMClassifier(
        n_estimators=1000,
        max_depth=8,
        learning_rate=0.03,
        num_leaves=70,
        min_child_samples=20,
        subsample=0.8,
        colsample_bytree=0.8,
        reg_alpha=0.3,
        reg_lambda=0.3,
        scale_pos_weight=scale_pos_weight,
        random_state=RANDOM_SEED,
        verbose=-1
    )
    lgb_final.fit(X_full, y_full)

    xgb_final = xgb.XGBClassifier(
        n_estimators=1000,
        max_depth=8,
        learning_rate=0.03,
        subsample=0.8,
        colsample_bytree=0.8,
        reg_alpha=0.3,
        reg_lambda=0.3,
        scale_pos_weight=scale_pos_weight,
        random_state=RANDOM_SEED,
        eval_metric='logloss',
        verbosity=0
    )
    xgb_final.fit(X_full, y_full)

    cat_final = CatBoostClassifier(
        iterations=1000,
        depth=8,
        learning_rate=0.03,
        l2_leaf_reg=5,
        scale_pos_weight=scale_pos_weight,
        random_state=RANDOM_SEED,
        verbose=False
    )
    cat_final.fit(X_full, y_full)

    final_model = {
        'type': 'ensemble',
        'base_models': {
            'LightGBM': lgb_final,
            'XGBoost': xgb_final,
            'CatBoost': cat_final
        },
        'weights': best_ensemble_config['weights'],
        'threshold': best_ensemble_config['threshold']
    }

elif best_holdout_model == 'Stacking_Ensemble':
    # Retrain base models + meta-learner for stacking
    print("Retraining base models and meta-learner for stacking ensemble...")

    lgb_final = lgb.LGBMClassifier(
        n_estimators=1000,
        max_depth=8,
        learning_rate=0.03,
        num_leaves=70,
        min_child_samples=20,
        subsample=0.8,
        colsample_bytree=0.8,
        reg_alpha=0.3,
        reg_lambda=0.3,
        scale_pos_weight=scale_pos_weight,
        random_state=RANDOM_SEED,
        verbose=-1
    )
    lgb_final.fit(X_full, y_full)

    xgb_final = xgb.XGBClassifier(
        n_estimators=1000,
        max_depth=8,
        learning_rate=0.03,
        subsample=0.8,
        colsample_bytree=0.8,
        reg_alpha=0.3,
        reg_lambda=0.3,
        scale_pos_weight=scale_pos_weight,
        random_state=RANDOM_SEED,
        eval_metric='logloss',
        verbosity=0
    )
    xgb_final.fit(X_full, y_full)

    cat_final = CatBoostClassifier(
        iterations=1000,
        depth=8,
        learning_rate=0.03,
        l2_leaf_reg=5,
        scale_pos_weight=scale_pos_weight,
        random_state=RANDOM_SEED,
        verbose=False
    )
    cat_final.fit(X_full, y_full)

    # Meta features for full data
    meta_full = np.column_stack([
        lgb_final.predict_proba(X_full)[:, 1],
        xgb_final.predict_proba(X_full)[:, 1],
        cat_final.predict_proba(X_full)[:, 1]
    ])
    meta_learner = LogisticRegression(C=1.0, class_weight='balanced', max_iter=1000, random_state=RANDOM_SEED)
    meta_learner.fit(meta_full, y_full)

    final_model = {
        'type': 'stacking',
        'base_models': {
            'LightGBM': lgb_final,
            'XGBoost': xgb_final,
            'CatBoost': cat_final
        },
        'meta_learner': meta_learner
    }

elif best_holdout_model == 'LightGBM':
    print("Retraining LightGBM on full data...")
    final_model = lgb.LGBMClassifier(
        n_estimators=1000,
        max_depth=8,
        learning_rate=0.03,
        num_leaves=70,
        min_child_samples=20,
        subsample=0.8,
        colsample_bytree=0.8,
        reg_alpha=0.3,
        reg_lambda=0.3,
        scale_pos_weight=scale_pos_weight,
        random_state=RANDOM_SEED,
        verbose=-1
    )
    final_model.fit(X_full, y_full)

elif best_holdout_model == 'XGBoost':
    print("Retraining XGBoost on full data...")
    final_model = xgb.XGBClassifier(
        n_estimators=1000,
        max_depth=8,
        learning_rate=0.03,
        subsample=0.8,
        colsample_bytree=0.8,
        reg_alpha=0.3,
        reg_lambda=0.3,
        scale_pos_weight=scale_pos_weight,
        random_state=RANDOM_SEED,
        eval_metric='logloss',
        verbosity=0
    )
    final_model.fit(X_full, y_full)

elif best_holdout_model == 'CatBoost':
    print("Retraining CatBoost on full data...")
    final_model = CatBoostClassifier(
        iterations=1000,
        depth=8,
        learning_rate=0.03,
        l2_leaf_reg=5,
        scale_pos_weight=scale_pos_weight,
        random_state=RANDOM_SEED,
        verbose=False
    )
    final_model.fit(X_full, y_full)

elif best_holdout_model == 'RandomForest':
    print("Retraining RandomForest on full data...")
    final_model = RandomForestClassifier(
        n_estimators=500,
        max_depth=15,
        min_samples_split=10,
        min_samples_leaf=5,
        class_weight='balanced',
        random_state=RANDOM_SEED,
        n_jobs=-1
    )
    final_model.fit(X_full, y_full)

else:
    # Fallback - use existing trained model if available, else default to LightGBM
    if best_holdout_model in models and hasattr(models[best_holdout_model], 'predict'):
        print(f"‚ö†Ô∏è Using existing trained model for {best_holdout_model}")
        final_model = models[best_holdout_model]
    else:
        print(f"‚ö†Ô∏è Unknown model '{best_holdout_model}', defaulting to LightGBM")
        final_model = lgb.LGBMClassifier(
            n_estimators=1000,
            max_depth=8,
            learning_rate=0.03,
            num_leaves=70,
            min_child_samples=20,
            subsample=0.8,
            colsample_bytree=0.8,
            reg_alpha=0.3,
            reg_lambda=0.3,
            scale_pos_weight=scale_pos_weight,
            random_state=RANDOM_SEED,
            verbose=-1
        )
        final_model.fit(X_full, y_full)

print("‚úÖ Final model retrained on full data!")

# B. Save artifacts
print("\nüíæ B. Saving model artifacts...")
print("-" * 40)

# Save final model
with open('best_model.pkl', 'wb') as f:
    pickle.dump(final_model, f)
print("‚úÖ best_model.pkl saved")

# Save feature extractor (already saved earlier in pipeline)
print("‚úÖ feature_extractor.pkl already saved")

# Save model configuration
model_config = {
    'best_model_name': best_holdout_model,
    'holdout_f1': float(best_holdout_f1),
    'holdout_precision': float(best_holdout_precision),
    'holdout_recall': float(best_holdout_recall),
    'ensemble_config': best_ensemble_config if isinstance(final_model, dict) and final_model.get('type') in ['ensemble', 'stacking'] else None,
    'scale_pos_weight': float(scale_pos_weight),
    'random_seed': int(RANDOM_SEED),
    'feature_count': int(X_train.shape[1]),
    'training_samples': int(len(X_full)),
    'target_achieved': bool(target_achieved)
}

with open('model_config.json', 'w') as f:
    json.dump(model_config, f, indent=2)
print("‚úÖ model_config.json saved")

# Save training metrics
training_metrics = []
for model_name, results in model_results.items():
    training_metrics.append({
        'model': model_name,
        'test_f1': float(results['f1_score']),
        'test_precision': float(results['precision']),
        'test_recall': float(results['recall']),
        'test_roc_auc': float(results['roc_auc'])
    })

# Merge holdout results if present
for model_name, results in holdout_results.items():
    for i, metric in enumerate(training_metrics):
        if metric['model'] == model_name:
            training_metrics[i].update({
                'holdout_f1': float(results['f1_score']),
                'holdout_precision': float(results['precision']),
                'holdout_recall': float(results['recall']),
                'holdout_roc_auc': float(results['roc_auc'])
            })

training_metrics_df = pd.DataFrame(training_metrics)
training_metrics_df.to_csv('training_metrics.csv', index=False)
print("‚úÖ training_metrics.csv saved")

print(f"\n‚úÖ All artifacts saved successfully!")
```
